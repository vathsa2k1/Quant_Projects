```python
def barplot_by_target(df_temp,x):
  variable_good_bad = df_temp.groupby(['Goodness_indicator',x]).size().reset_index(name='count')
  title = "{} by Good_Bad Customers".format(x)
  fig = px.bar(variable_good_bad, x=x, y="count", color="Goodness_indicator", title=title)
  fig.show()
```

```python
barplot_by_target(df_data,'term')
```

```python
barplot_by_target(df_data,'term')
```
